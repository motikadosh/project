```python
import random
import numpy as np
import meshNet_loader
```

```python
data_dir = '/home/moti/cg/project/sessions_outputs/project_2017_07_30-11_27_19-300sampelsMin5LinesUpper0.33'
loader = meshNet_loader.DataLoader(
        data_dir=data_dir, part_of_data=1000)
```

```python
y = np.concatenate((loader.y_train, loader.y_test))
file_urls = np.concatenate((loader.file_urls_train, loader.file_urls_test))

min_x = np.min(y[:, 0])
max_x = np.max(y[:, 0])
min_y = np.min(y[:, 1])
max_y = np.max(y[:, 1])
print (min_x)
print (max_x)
print (min_y)
print (max_y)
```

```python
# Create totally random predictions
# Notice y is normalized to [0, 1], i.e. its min/max are in [0, 1] so I just use that hard coded
y_pred = np.random.uniform(0, 1, (y.shape[0], 4))
print(y_pred.shape)
print(y_pred.mean())
```

```python
def detailed_evaluation(y, y_pred, file_urls, loader):
    print("detailed evaluation...")

    # Get Y values de-normalized
    y = loader.y_inverse_transform(y)
    y_pred = loader.y_inverse_transform(y_pred)
    
    xy_error = np.linalg.norm(y[:, 0:2] - y_pred[:, 0:2], axis=-1)
    xy_error_mean = np.mean(xy_error)
    xy_error_std = np.std(xy_error)
    
    # Print results
    for i in xrange(y_pred.shape[0]):
        print("Sample #%s. File: %s" % (i, file_urls[i]))
        print ("Ground-Truth: %s, Prediction: %s, xy_error: %s" % (y[i], y_pred[i], xy_error[i]))
        print ("")
    
    print("Average xy_error: %s" % xy_error_mean)
    print("Standard deviation xy_error: %s" % xy_error_std)
```

```python
detailed_evaluation(y, y_pred, file_urls, loader)
```
